## Carga de Audio

```python
import os

VOSTOK_ROOT = "/content/drive/MyDrive/desarrollos/vostok restoration v1/Vostok-ML-Research-Lab"

RAW_DATASET = f"{VOSTOK_ROOT}/datasets/raw"

print(os.listdir(RAW_DATASET))
```

```python
audio_path = f"{RAW_DATASET}/vozenoff.wav"

y, sr = librosa.load(audio_path, sr=None)

print(f"Sample Rate: {sr}")
print(f"Muestras: {len(y)}")
print(f"Duración: {len(y)/sr:.2f} segundos")
```

## Visualización de Forma de Onda

```python
plt.figure(figsize=(15,4))
plt.plot(y)
plt.title("vozenoff.wav")
plt.xlabel("Muestras")
plt.ylabel("Amplitud")
plt.show()
```

## Espectrograma STFT

```python
D = librosa.amplitude_to_db(
    np.abs(librosa.stft(y)),
    ref=np.max
)

plt.figure(figsize=(15,8))

librosa.display.specshow(
    D,
    sr=sr,
    x_axis='time',
    y_axis='log'
)

plt.colorbar(format='%+2.0f dB')
plt.title("Espectrograma STFT - vozenoff.wav")

plt.show()
```
